```python
#  Send E-mail multiple users with attachments

import smtplib
from email.mime.multipart import MIMEMultipart
from email.mime.text import MIMEText
from email.mime.application import MIMEApplication
from email.mime.image import MIMEImage
from email.mime.audio import MIMEAudio
import pandas as pd  # type: ignore

# SMTP Configuration
smtp_server = 'smtp.gmail.com'
smtp_port = 587
smtp_username = 'agrawaluma2002@gmail.com'
smtp_password = 'jdjs yfvb rcyw qzlh'

# File Paths
file_path = "/content/drive/MyDrive/filtered_emails42.xlsx"
attachment_paths = ["/content/drive/MyDrive/DOC-20250213-WA0030..pdf",
                    "/content/drive/MyDrive/sukhdevji2.jpg",
                    "/content/drive/MyDrive/WhatsApp Image 2025-02-19 at 11.41.36_83c09d83.jpg"]


# Read Excel File
df = pd.read_excel(file_path)
def get_email_body(recipient_name):
    return f"""Dear {recipient_name},
"Dear Members of 𝐈𝐍𝐓𝐀𝐂𝐇, we are currently engaged in a contest for the position of Chairman and kindly seek your attention and support.",

WHAT HE THINKS OF INTACH: as an Organisation & an Empowering Citizen's body

𝐈𝐍𝐓𝐀𝐂𝐇
• To think and act in interests of society
• To think globally but act locally
• To work as a body of aware, active & passionate members
• To work above personal interests, ambitions and ego

WHAT HE PROPOSES FOR ACTUAL POSITIVE CHANGES

𝐓𝐫𝐮𝐞 𝐬𝐮𝐩𝐩𝐨𝐫𝐭 & 𝐩𝐨𝐰𝐞𝐫 𝐟𝐨𝐫 𝐜𝐡𝐚𝐩𝐭𝐞𝐫𝐬
• Biilding cordial relations with central and state governments
• INTACH in UNESCO & other fora
• Membership audit & comprehensive correction
• To bring transparency in management & information sharing
• To limit terms for the Governing Council Members, Chairman & Member Secretary
• Financial and Technical Capacity Building of Chapters
• Interactive platforms for MEMBERS, CONVENERS & GOVERNING COUNCIL
• Human Resource Policy
• Active involvement of institutions as knowledge partners
• Active dialoque with corporate on CSR

𝐏𝐋𝐄𝐀𝐒𝐄 𝐕𝐎𝐓𝐄 & 𝐒𝐔𝐏𝐏𝐎𝐑𝐓
𝐏𝐑𝐎𝐅 𝐒𝐔𝐊𝐇𝐃𝐄𝐕 𝐒𝐈𝐍𝐆𝐇
𝐟𝐨𝐫 𝐂𝐇𝐀𝐈𝐑𝐌𝐀𝐍, 𝐈𝐍𝐓𝐀𝐂𝐇 𝐈𝐍𝐃𝐈𝐀 (𝐁𝐇𝐀𝐑𝐀𝐓)
             𝐂𝐚𝐧𝐝𝐢𝐝𝐚𝐭𝐞 𝐍𝐨 𝟑
"""
# Iterate through rows and send emails
for index, row in df.iterrows():
    recipient_name = row['First Name']
    recipient_email = row['E-mail (Primary)']

    if pd.isna(recipient_email) or pd.isna(recipient_name):
        print(f"Skipping row {index}: Missing email or name.")
        continue


    body = get_email_body(recipient_name)

    # Create Message
    msg = MIMEMultipart()
    msg['From'] = smtp_username
    msg['To'] = recipient_email
    msg['Subject'] = "A New Era for INTACH: Support Sukhdev Ji for Chairman!"
    msg.attach(MIMEText(body))

    # Attach File
    for attachment_path in attachment_paths:
        try:
            if attachment_path.lower().endswith('.pdf'):
                with open(attachment_path, 'rb') as f:
                    attachment = MIMEApplication(f.read(), _subtype='pdf')
                    attachment.add_header('Content-Disposition', 'attachment', filename='Sukhdevji.pdf')
                    msg.attach(attachment)
            elif attachment_path.lower().endswith(('.jpg','.jpeg','.png')):
                with open(attachment_path, 'rb') as f:
                    attachment = MIMEImage(f.read())
                    attachment.add_header('Content-Disposition', 'attachment', filename='Image.png')
                    msg.attach(attachment)
        except FileNotFoundError:
            print(f"Attachment file not found: {attachment_path}. Email will be sent without the attachment.")


    # Send Email
    try:
        with smtplib.SMTP(smtp_server, smtp_port) as smtp:
            smtp.starttls()
            smtp.login(smtp_username, smtp_password)
            smtp.send_message(msg)
            print(f"Email successfully sent to {recipient_name} ({recipient_email})")
    except smtplib.SMTPException as e:
        print(f"Failed to send email to {recipient_name} ({recipient_email}). Error: {e}")


```
